```python
import time
import json
import requests
import pandas as pd
import googlemaps
gmaps = googlemaps.Client(key = 'Your API KEY')
```

```python
#AIzaSyCUUvP-TGxfZ0nqMOEcs7PmpET_ZeyeYpk
def NearbySeach(key, location, radius,type1):
    """
    Function to obtain data of places which are near to a specific location using request and json

    Args:
        key: our API key which will give us the access of the Google's API
        location: Where is the place that we want to know as LAT/LONG in string separated by commas
        radius: A radius in meters to know how near are the places
        type1:
        
    Returns:

     A dictionary with the places which are near in the given radius


    """    
    

    endpoint_url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?'
    params = {
        'key' : key,
        'location' : location,
        'radius' : radius,
        'type' : type1
    }
    
    res = requests.get(endpoint_url, params= params)
    results = res.content
    return results
```

B. Create a function to Find the Place Id given Place Name (String) and Location (Lat/Lont Tuple).

```python
def idPlace(name, key, location):
    """
    Function to obtain the ID of a place given the name and the location

    Args:
        name: The name of the place
        key: Our API key to access to the Google's API
        location: Where is the place that we want to know as LAT/LONG in string separated by commas

    Returns:
        
     A dictionary with the ID of the place
     

    """
    
    endpoint_url = 'https://maps.googleapis.com/maps/api/place/findplacefromtext/json?'
    locationbias = 'circle:100@'+ location
    params = {
        'input' : name,
        'key' : key,
        'inputtype' : 'textquery',
        'locationbias' : locationbias,
    }
    
    res = requests.get(endpoint_url, params = params)
    results = res.content
    results = json.loads(results)
    return results
```

```python
def find_request(key, iD, language):
    """
    Function to obtaion the reviews of a Place in a specific language

    Args:
        key: Our API key to access to the Google's API
        iD: The ID of the place
        language: The language of the reviews

    Returns:
        
     A dictionary with the reviews in the given language
     

    """

    endpoint_url = 'https://maps.googleapis.com/maps/api/place/details/json?'
    params = {
        'key' : key,
        'place_id' : iD,
        'language' : language
    }
    
    res = requests.get(endpoint_url, params = params)
    results = res.content
    results = json.loads(results)
    return results
```

E. Find similar places to Los Trompos Circuito. Rank them by its "Prominence". In order to fulfill this item you need to:
* Create a *find_similar* function with the appropiate parameters. It must return a JSON with the results.
* Create a table with the following columns: Place Name, Plaece Id, Rating, Total of Use Ratings.

```python
def find_similar(key, location, radius,type1):
    """
    Function to obtain similar places of a given place

    Args:
        key: Our API key to access to the Google's API
        location: Where is the place that we want to know as LAT/LONG in string separated by commas
        radius: The radius of how long we want to cover
        type1: the type of place that we want to search (restaurant, bar, etc)

    Returns:
        
     A dictionary with the places which are similar
     

    """
    endpoint_url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?'
    params = {
        'key' : key,
        'location' : location,
        'radius' : radius,
        'type' : type1,
        'rankby' : 'prominence'
    }
    
    res = requests.get(endpoint_url, params= params)
    results = res.content
    results = json.loads(results)
    return results
```

```python
similarTrompos = find_similar('Your API KEY', '20.999070, -89.617521', 2000,'restaurant')
similarTrompos = pd.read_json(json.dumps(similarTrompos['results']), orient='records')
similarTrompos.drop(["geometry","id", "icon", "opening_hours", "photos","types", "plus_code", "price_level", "reference","vicinity","scope"], axis=1, inplace=True)
similarTrompos.head()
```
